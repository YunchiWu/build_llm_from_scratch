# 1. 理解大语言模型
## 1.1 什么是大语言模型
大语言模型是一种用于理解、生成和响应类似人类语言文本的神经网络。大语言模型中的“大”既体现了模型训练所依赖的庞大数据集，
又反映了模型本身庞大的参数规模。大语言模型通常拥有数百亿（10B）甚至数千亿（100B）个参数。这些参数是神经网络中的可调整权重，在训练过程中不断被优化。

由于大语言模型能够生成文本，因此它们通常也被归类为**生成式人工智能（Generative AI，GenAI）**。具体而言，人工智能旨在开发完成需要人类智力水平的任务的机器；
机器学习研究实现人工智能的算法，聚焦能够从数据中学习的算法；深度学习是机器学习的分支，利用神经网络建模数据中的复杂模式和抽象特征。

因此，人工智能、机器学习、深度学习、大语言模型以及GenAI的关系是：
人工智能 > 机器学习 > 深度学习 > 大语言模型，其中GenAI是深度学习和大语言模型技术的一种。

## 1.2 大语言模型的应用
大语言模型擅长解析和理解非结构化文本数据，目前被广泛用于机器翻译、文本生成、情感分析、文本摘要等多种任务。
大语言模型在需要解析和生成文本的任务的自动化处理中具有重要价值。

# 1.3 构建和使用大语言模型的各个阶段
从零开始构建大模型是一次深入了解模型机制和局限性的绝佳机会，还提供了预训练和微调开源大模型，使其适应特定领域的数据集或任务的必要知识。

使用定制的大模型有多个优势，尤其是在数据隐私方面。例子有三：第一，出于机密性考虑，敏感数据可能不适合共享给OpenAI等第三方大语言模型提供商；
第二，如果开发较小的定制化大语言模型，那么就可以直接部署到自己的设备（笔记本电脑或智能手机）上，显著降低延迟以及与服务器相关的成本；第三，
定制的大语言模型使得开发者拥有完全的自主权，能够根据需要控制模型的更新和修改。

大语言模型的构建通常包括预训练（pre-training）和微调（fine-tuning）两个阶段。“预训练”中的“预”表明它是模型训练的初始阶段，
此时模型会在大规模、多样化的数据集上进行训练。以预训练阶段为基础，微调阶段会在规模较小的特定任务或领域数据集上对模型进行针对性训练，以进一步提升模型的特定能力。
综上，大语言模型的预训练目标是在大量无标注的文本语料库（原始文本，raw text）上进行下一单词预测（采用自监督学习技术，模型从输入数据中产生自己的标签）；
预训练完成后，可以使用较小的带标注数据集对大语言模型进行微调。

预训练是大语言模型的第一个训练阶段，预训练后的大语言模型称为基础模型（foundation model），一个例子是ChatGPT的前身GPT-3，其具有文本补全能力和少样本学习能力。
随后，在带标注的数据集上进一步训练基础模型，这一阶段称之为微调。微调大语言模型的两种方式是指令微调（instruction fine-tuning）和分类任务微调（classification fine-tuning）。
指令微调中标注数据集由“指令-答案”对组成（比如翻译任务中的“原文-正确翻译文本”）；分类任务微调中标注数据集由文本及其类别标签组成（比如已被标记为“垃圾邮件”或“非垃圾邮件”的电子邮件文本）。
